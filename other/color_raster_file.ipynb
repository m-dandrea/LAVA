```python
# Create colormap
sorted_items = sorted(color_dict.items())
positions, colors = zip(*sorted_items)
cmap = LinearSegmentedColormap.from_list("wind_speed_cmap", list(zip(positions, colors)))

# Load raster
raster_path = os.path.join('..', 'data', "Elbe-Elster", 'wind_Elbe-Elster_EPSG32633.tif')
with rasterio.open(raster_path) as src:
    wind_speed = src.read(1)
```

```python
plt.imshow(wind_speed)
```

```python
# Normalize only values < 10 by dividing by 10, else assign 1
normed_wind_speed = np.where(wind_speed < 10, wind_speed / 10, 1.0)

# Apply colormap
colored_wind_speed = cmap(normed_wind_speed)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(8,6))

img = ax.imshow(cmap(normed_wind_speed))

# Create ScalarMappable for colorbar with normalization from 0 to 10
norm = Normalize(vmin=0, vmax=10)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Dummy mappable to connect colorbar

cbar = fig.colorbar(sm, ax=ax, label='Wind Speed (m/s)')
plt.title("Colored Wind Speed Map\n(Values >=10 shown with last color)")
plt.show()
```

```python
# Convert float RGBA [0..1] to uint8 [0..255]
rgba_uint8 = (colored_wind_speed * 255).astype(np.uint8)

# Rearrange from H x W x 4 to 4 x H x W for rasterio
rgba_uint8 = np.transpose(rgba_uint8, (2, 0, 1))

# Prepare metadata for output
meta = src.meta.copy()
meta.update({
    "count": 4,            # RGBA has 4 bands
    "dtype": "uint8",
    "nodata": None         # NoData can be None for RGBA
})

# Output filename
output_tif = "colored_wind_speed.tif"

with rasterio.open(output_tif, 'w', **meta) as dst:
    dst.write(rgba_uint8)
```
